```python
# change values to integer

df_clean = df.copy()
df_clean.price = df_clean.price.astype(int)
df_clean.sqft_living15 = df_clean.sqft_living15.astype(int)
df_clean.sqft_lot15 = df_clean.sqft_lot15.astype(int)
df_clean.sqft_living = df_clean.sqft_living.astype(int)
df_clean.sqft_lot = df_clean.sqft_lot.astype(int)
df_clean.bedrooms = df_clean.bedrooms.astype(int)
df_clean.bathrooms = df_clean.bathrooms.astype(int)
df_clean.floors = df_clean.floors.astype(int)
```

```python
# drop columns we don't need

df_clean.drop(["house_id", "yr_renovated", ], axis=1, inplace=True)
df_clean.head()
```

```python
# due to his limited budget => focus on prices up to the median = $ 450.000

median_110 = int(df_clean['price'].median() * 1.1)
median_110
```

```python
# Calculate the average house price per zipcode
mean_price_per_zipcode = df_clean.groupby('zipcode')['price'].mean().reset_index()
mean_price_per_zipcode.columns = ['zipcode', 'average_price']
mean_price_per_zipcode
```

```python
# filter data into new dataframe

# Step 1: Filter houses with a price lower than the median + 10%
median_price = df_clean['price'].median()
df_filter = df_clean[df_clean['price'] < (median_price * 1.5)]

# Step 2: Filter houses with sqft_lot15 higher than the median
median_sqft_lot15 = df_filter['sqft_lot15'].median()
df_filter = df_filter[df_filter['sqft_lot15'] > median_sqft_lot15]

# Step 3: Filter houses with sqft_living15 higher than the median
median_sqft_living15 = df_filter['sqft_living15'].median()
df_filter = df_filter[df_filter['sqft_living15'] > median_sqft_living15]

# Step 4: Filter houses with waterfront = 1
df_filter = df_filter[df_filter['waterfront'] == 1]

# Step 5: Sort by grading from high to low (grade)
df_filter = df_filter.sort_values(by='grade', ascending=False)

# Step 6: Filter houses with at least 3 bedrooms
df_filter = df_filter[df_filter['bedrooms'] >= 3]

# Step 7: Filter houses with at least 2 bathrooms
df_filter = df_filter[df_filter['bathrooms'] >= 1]

# Display the cleaned DataFrame
print(df_filter)
```

```python
import folium

# Create a map centered around King County
m = folium.Map(location=[47.6062, -122.3321], zoom_start=10)

# Iterate through the DataFrame and add markers for each house
for index, row in df_filter.iterrows():
    folium.Marker(
        location=[row['lat'], row['long']],
        popup=f"Price: ${row['price']:.2f}",
        icon=folium.Icon(color='blue')
    ).add_to(m)

# Display the map
m.save('king_county_houses_map.html')

```
